```python
spark.sql("CREATE SCHEMA IF NOT EXISTS prata")
spark.sql("SELECT * FROM bronze.despesas").createOrReplaceTempView("bronze_despesas_view")
print("Pronto.")
```

# Tabela fato_despesas
Tabela prata com dimensões: deputado, fornecedor, categoria, mês.

```python
%sql
CREATE OR REPLACE TABLE prata.fato_despesas
USING DELTA
COMMENT 'Camada Prata - Fato de despesas da CEAP com dimensoes'
AS
SELECT 
    id_deputado,
    nome_deputado,
    sigla_partido,
    sigla_uf,
    ano,
    mes,
    tipo_despesa AS categoria,
    CAST(data_documento AS DATE) AS data_documento,
    valor_documento,
    valor_liquido,
    valor_glosa,
    nome_fornecedor,
    cnpj_cpf_fornecedor,
    url_documento,
    data_coleta,
    origem
FROM bronze_despesas_view
```

```python
%sql
SELECT 
    *
FROM prata.fato_despesas
limit 10
```
