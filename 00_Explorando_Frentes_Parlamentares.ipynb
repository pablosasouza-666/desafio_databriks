```python
# ============================================================
# EXPLORANDO O ENDPOINT DE FRENTES PARLAMENTARES
# ============================================================
import requests
import json

# Endpoint de frentes
url = "https://dadosabertos.camara.leg.br/api/v2/frentes"

# Parâmetros CORRIGIDOS (removidos ordem e ordenarPor)
params = {
    "itens": 5  # Só 5 para explorar
}

response = requests.get(url, params=params)
print(f"Status: {response.status_code}")

if response.status_code == 200:
    dados = response.json()
    # Mostrar total de registros (se disponível)
    print(f"Metadados: {json.dumps(dados.get('dados', []), indent=2, ensure_ascii=False)[:500]}")
    print("\n" + "="*50)
    print(json.dumps(dados, indent=2, ensure_ascii=False)[:3000])
else:
    print(f"Erro: {response.text}")
```

```python
# Pegar a última página para ver quantos itens exatos tem
url = "https://dadosabertos.camara.leg.br/api/v2/frentes"
params = {
    "itens": 5,
    "pagina": 288  # Última página que descobrimos
}

response = requests.get(url, params=params)
if response.status_code == 200:
    dados = response.json()
    total_itens_ultima = len(dados['dados'])
    total_exato = (287 * 5) + total_itens_ultima
    print(f"Itens na última página: {total_itens_ultima}")
    print(f"Total EXATO de frentes: {total_exato}")
```
