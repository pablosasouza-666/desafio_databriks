## Desafios Implementados

### 1. Atlas das Frentes Parlamentares
- 1.440 frentes mapeadas com 261.637 membros
- Índice de Herfindahl (diversidade partidária)
- Deputados com mais frentes e sobreposição ideológica
- Evolução de frentes por tema entre legislaturas

### 2. Calendário Analítico de Eventos
- 11.022 eventos de 2022 a 2025
- Taxa de presença por deputado e tipo de evento
- Comparativo antes/depois de eleições (2024)
- Densidade semanal e semanas sem atividade
- View de calendário de eventos futuros

### 3. Correlação Frentes × Votações
- Score ideológico dos deputados via frentes (proxy)
- Comparativo de coesão: partidos (0,324) vs frentes (0,338)
- Partidos são ligeiramente mais coesos

### 4. Raio-X de Gastos da CEAP
- 341.586 despesas de 560 deputados (2024)
- R$ 384,3 milhões em reembolsos analisados
- Z-Score de anomalias por categoria × estado
- Ranking de fornecedores e top 10 gastos por partido

### 5. Pipeline de Auditoria de CPIs
- 4 CPIs analisadas (Criptomoedas, Futebol, MST, Americanas)
- Timeline com 126 eventos e 3.646 participações
- Requerimentos de criação vinculados a cada CPI
- Análise de duração e produtividade

### 6. Monitor de Presença e Absenteísmo
- Score de engajamento (0-100) para 564 deputados
- Padrão de ausência por tipo de evento
- Série temporal com queda no período eleitoral (set/2024)
- Relatório mensal com percentil

## Decisões Técnicas

- **Padrão medalhão:** cada camada com propósito claro (imutabilidade, limpeza, analytics)
- **Snake case:** padronização de nomes de colunas em todas as tabelas
- **Paginação:** coleta completa de todos os endpoints, com pausa entre chamadas
- **Amostra vs coleta completa:** despesas coletadas para todos os deputados; presenças apenas para 2024 (ano eleitoral mais recente)
- **Classificação ideológica:** baseada em palavras-chave nos títulos das frentes

## Limitações e Pendências

| Item | Motivo |
|------|--------|
| API de votações | Erro 500 em todas as tentativas |
| Classificação ideológica | Proxy via palavras-chave, não dados reais de voto |
| Relatório automatizado | Community Edition não suporta jobs agendados |
| Rede de convocados de CPI | API não oferece endpoint específico |
| Presenças | Apenas 2024 (ano eleitoral); demais anos não coletados |
